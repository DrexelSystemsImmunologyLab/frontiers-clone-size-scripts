```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

matplotlib.style.use('ggplot')
```

```python
for cutoff in range(1, 6):
    df = pd.DataFrame.from_dict({
        metric: pd.read_csv('similarity_C{}_{}.tsv'.format(cutoff, metric), sep='\t')['similarity']
        for metric in ('copies', 'instances')
    })
    ax = df.plot.box(figsize=(12, 8), showfliers=False)
    ax.set_title(cutoff)
    ax.set_ylim((0, 1))
    plt.savefig('cosine_C{}.pdf'.format(cutoff), bbox_inches='tight')
```
